### Customizing plots

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,4*np.pi,100)
plt.plot(x, np.sin(x))
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('The sin function')
plt.ylim(-1.5,1.5)
plt.grid(True)
```

### Adding a legend

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,4*np.pi,100)

plt.plot(x, np.sin(x), "r", label="sine")
plt.plot(x, np.cos(x), "b", label="cosine")
plt.xlabel('x')
plt.ylabel('sin (x)/cos(x)')
plt.title('The sin and cos function')
plt.legend()
```

### Using subplots

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 4 * np.pi, 100)
plt.subplot(1,2,1)
plt.plot(x, np.sin(x), "r")
plt.xlabel("x"); plt.ylabel("sin(x)"); plt.title("The sine function")
plt.subplot(1,2,2)
plt.plot(x, np.cos(x), "b")
plt.xlabel("x"); plt.ylabel("cos(x)"); plt.title("The cosine functions")
plt.subplots_adjust(wspace=0.5)
```

### Q1. Plot the tangent function of values from 0 to 4π, and using 1000 data points. Specify the third parameter of plot() as “r:” (see the result for what this does). Set the range of y-values from –10 to 10, and use proper labels and title.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 data points from 0 to 4π
x = np.linspace(0, 4 * np.pi, 1000)
y = np.tan(x)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'r:')  # 'r:' specifies a red dotted line
plt.ylim(-10, 10)  # Set the range of y-values from -10 to 10
plt.title('Tangent Function from 0 to 4π')
plt.xlabel('x (radians)')
plt.ylabel('tan(x)')
plt.axhline(0, color='black',linewidth=0.5, ls='--')  # Add x-axis
plt.axvline(0, color='black',linewidth=0.5, ls='--')  # Add y-axis
plt.grid(True)

# Show the plot
plt.show()
```

### (d) Using the df DataFrame above, write code to plot the data using the x Series as the x-values of the plot.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Define the DataFrame
x = np.linspace(0, 6 * np.pi, 1000)
df = pd.DataFrame({
    "x": x,
    "sin": np.sin(x),
    "cos": np.cos(x),
    "log10": np.log10(x, where=x > 0)
})

# Create the plot
plt.figure(figsize=(12, 6))

# Plot sine
plt.plot(df['x'], df['sin'], label='sin(x)', color='blue')

# Plot cosine
plt.plot(df['x'], df['cos'], label='cos(x)', color='orange')

# Plot logarithm
plt.plot(df['x'], df['log10'], label='log10(x)', color='green')

# Add titles and labels
plt.title('Plot of sin(x), cos(x), and log10(x)')
plt.xlabel('x (radians)')
plt.ylabel('Value')
plt.ylim(-1.5, 2)  # Adjust y-limits for better visibility
plt.axhline(0, color='black', linewidth=0.5, ls='--')  # Add x-axis
plt.axvline(0, color='black', linewidth=0.5, ls='--')  # Add y-axis
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
